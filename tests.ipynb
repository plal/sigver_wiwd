```python
import numpy as np
import os
import pandas as pd
from scipy.io import loadmat
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
features = loadmat('gpds_signet/exploitation/forg_2.mat')['features']
features
```

```python
data = []
path = "gpds_signet/exploitation"
for filename in os.listdir(path):
    if "real" in filename:
        features = loadmat(path + '/' + filename)['features']
        for f in features:
            f_ = np.append(f,1)
            data.append(f_)
    if "forg" in filename:
        features = loadmat(path + '/' + filename)['features']
        for f in features:
            f_ = np.append(f,0)
            data.append(list(f_))
            
print(len(data))
```

```python
labels = []
for i in range(1,2049):
    labels.append(str(i))
labels.append("real_or_forg")
print(len(labels))
```

```python
df = pd.DataFrame(data=data, columns=labels)
df
```

```python
df.to_csv('test_df.csv')
```

```python
data_ = pd.read_csv('test_df.csv')

data2 = data_.drop('real_or_forg', axis=1)
data2
```

```python
target = data_['real_or_forg']
target
```

```python
data_.loc[data_['real_or_forg'] == 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data2, target, test_size=0.3, random_state=0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = svm.SVC(kernel='rbf', gamma = 1/2048, C=1).fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, random_state=0)
for train_index, test_index in kf.split(data2, target):
    X_train2, X_test2 = data2.iloc[train_index], data2.iloc[test_index]
    y_train2, y_test2 = target.iloc[train_index], target.iloc[test_index]
    classif = svm.SVC(kernel='rbf', gamma = 1/2048, C=1).fit(X_train2, y_train2)
    print(classif.score(X_test2, y_test2))
```
